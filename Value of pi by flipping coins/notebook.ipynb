```python
import numpy as np
import matplotlib.pyplot as plt
```

### Procedure

- We start with `sum = 0`.
- The coin is tossed `steps` times. 
- Each time, if the outcome is head, `sum = sum + 1`, if tails, `sum = sum - 1`.
- The absolute value of sum `sum = np.abs(sum)` is stored in an `array`

These above steps are done `runs` times and `array` is updated each time.

The mean `m` of `array` is found `m = np.mean(array)` and the value of pi is given by the equation $Ï€ = \frac{2n}{m^2}$

Now all of the above steps are performed with `steps = steps + 5` upto `steps = 1000`

```python
start_step = 5

end_step = 1000

step_array = np.arange(start_step, end_step+1, 5, dtype='int16')
pi_value_array_calculated = np.zeros(len(step_array), dtype='float32')
pi_value_array = np.full(len(step_array), np.pi, dtype='float32')

runs = 1000000

for i in range(len(step_array)):
    pi_value_array_calculated[i] = (2*step_array[i])/np.mean(np.abs(np.sum(np.random.choice([np.int8(-1), np.int8(1)], size=(runs, step_array[i])), axis=1, dtype='int16')))**2

plt.figure(figsize=(30, 7))
plt.plot(step_array, pi_value_array, 'r', linewidth=2)
plt.plot(step_array, pi_value_array_calculated)
plt.xlabel('number of tosses')
plt.title('Approximating pi with coin toss')
plt.show()

print('AVERAGE VALUE OF ALL VALUES CALCULATED', np.mean(pi_value_array_calculated))
```
